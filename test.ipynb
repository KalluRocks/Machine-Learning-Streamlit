```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
height = np.random.normal(loc=5.5, scale=0.5/3, size=10000)
state = np.random.choice(['Maharashtra', 'Bihar', 'UP', 'Delhi', 'Manipur'], size=10000)
```

```python
height
```

```python
state
```

```python
dct = {'State':state,
       'Height':height}
```

```python
df = pd.DataFrame(dct)
df.head()
```

```python
df.shape
```

```python
df['Height'].describe()
```

```python
df['State'].describe()
```

```python
df['State'].value_counts()
```

```python
df.groupby(by='State').median()['Height']
```

```python
sns.histplot(data=df, x='Height', hue='State', kde=True)
```
